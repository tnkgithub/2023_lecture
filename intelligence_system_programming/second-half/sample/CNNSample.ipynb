```python
# 必要なライブラリをインポート
import torch
from torchvision import models, transforms

from PIL import Image

import json
```

```python
## torchvision.modelsからImageNet学習済みモデルをロード
#model_image = models.vgg16(pretrained=True)
#model_image = models.vgg19(pretrained=True)
#model_image = models.alexnet(pretrained=True)
model_image = models.resnet152(pretrained=True)

##学習済みモデルを評価モードに
model_image.eval()
```

```python
##ImageNetの画像に合わせて入力画像を変形させる
image_converter = transforms.Compose([
    transforms.Resize(224),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])
```

```python
##入力画像を指定
target_image = Image.open('images/car.jpg')
##入力画像を変形
converted_image = image_converter(target_image)
```

```python
#バッチデータとして形式を合わせる
batch = converted_image[None]
```

```python
#画像の推論
result = model_image(batch)
```

```python
#推論結果を取得
result_index = torch.argmax(result[0])
```

```python
#推論結果のカテゴリ番号から判定した対象名を取得，表示
with open('./imagenet_class_index.json') as file:
    labels = json.load(file)
    answer = labels[str(result_index.item())]
    print(answer)
```
